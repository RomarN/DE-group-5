```python
def loadcsv(df): 
    df1 = df
    #Convert the categorical data into binary values
    df_bin = df1.join(pd.get_dummies(df1.pop('day'))) 
    df_bin = df_bin.join(pd.get_dummies(df_bin.pop('month')))
    df_bin.pop('sun')
    try:
        df_bin.pop('dec')
    except:
        return df_bin
    #df_bin = df_bin[df_bin['area'] < 3.2]  
    # df_bin.loc[df.area > 0, 'area']= 1
    #Return the database
    return df_bin
```
